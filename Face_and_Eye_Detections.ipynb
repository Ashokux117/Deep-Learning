let's make a live fece and eye detection kipping the face inview at all  time

```python
import cv2
import numpy as np

# Load Haar Cascade Classifiers
eyes_classifier = cv2.CascadeClassifier(r"C:\Users\DELL\OneDrive\Documents\haarcascade_eye.xml")
face_classifier = cv2.CascadeClassifier(r"C:\Users\DELL\OneDrive\Documents\haarcascade_frontalface_default.xml")

def face_detector(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_classifier.detectMultiScale(gray, 1.3, 5)
    
    if len(faces) == 0:
        print("No faces found")
        return img
    
    for (x, y, w, h) in faces:
        cv2.rectangle(img, (x, y), (x + w, y + h), (127, 0, 255), 2)
        
        roi_gray = gray[y:y + h, x:x + w]
        roi_color = img[y:y + h, x:x + w]
        
        eyes = eyes_classifier.detectMultiScale(roi_gray)
        for (ex, ey, ew, eh) in eyes:
            cv2.rectangle(roi_color, (ex, ey), (ex + ew, ey + eh), (255, 255, 0), 2)
    
    return img

# Start webcam
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break
    
    frame = face_detector(frame)
    cv2.imshow("Face & Eye Detection", frame)
    
    # Press 'q' to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
